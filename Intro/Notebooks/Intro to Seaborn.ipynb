# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot

## Data
Seaborn comes with built-in data sets!

````python
tips = sns.load_dataset('tips')
````

````python
tips.head()
````

## distplot

The distplot shows the distribution of a univariate set of observations.

````python
sns.distplot(tips['total_bill'],kde=True);
````

To remove the kde layer and just have the histogram use:

````python
sns.distplot(tips['total_bill'],kde=False,bins=20);
````

## jointplot

jointplot() allows you to match up two distplots for bivariate data.

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter');
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',size=7, space=0.7);
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg');
````

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns)

````python
sns.pairplot(tips);
````

````python
sns.pairplot(tips,hue='smoker',palette='coolwarm')
````

# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

## countplot

````python
sns.countplot(x='sex',data=tips)
````

## boxplot

````python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
````

````python
sns.boxplot(data=tips,palette='rainbow',orient='h')
````

## heatmap

````python
# Matrix form for correlation data
tips.corr()
````

````python
sns.heatmap(tips.corr())
````

## pairplot

````python
sns.pairplot(tips)
````

````python
sns.pairplot(tips, hue='sex')
````
